3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculates the PDF (PMF) of a Bernoulli distribution

    Parameters:
    x : int (0 or 1)
    p : float (probability of success, 0 <= p <= 1)

    Returns:
    float : probability value
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

```python
import matplotlib.pyplot as plt
import numpy as np

# Discrete uniform distribution values
values = np.array([1, 2, 3, 4, 5, 6])

# PMF (equal probability)
pmf = np.ones(len(values)) / len(values)

# CDF (cumulative sum of PMF)
cdf = np.cumsum(pmf)

# Plot CDF
plt.step(values, cdf, where='post')
plt.xlabel("Random Variable (X)")
plt.ylabel("CDF: P(X â‰¤ x)")
plt.title("CDF of Discrete Uniform Distribution")
plt.xticks(values)
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset

    Parameters:
    data : list or array of numerical values

    Returns:
    z_scores : array of Z-scores
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Sample dataset
data = np.array([45, 50, 55, 60, 65, 70, 75])

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot Z-scores
plt.bar(range(len(z_scores)), z_scores)
plt.xlabel("Data Index")
plt.ylabel("Z-score")
plt.title("Z-scores of the Dataset")
plt.show()
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 100000
sample_size = 30        # n â‰¥ 30 for CLT
num_samples = 1000      # number of sample means

# Generate a non-normal population (Exponential distribution)
population = np.random.exponential(scale=1.0, size=population_size)

# Draw samples and compute sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True)
plt.xlabel("Sample Mean")
plt.ylabel("Probability Density")
plt.title("Central Limit Theorem (Exponential Distribution)")
plt.show()
```

11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

```python
import numpy as np
import matplotlib.pyplot as plt

# Population parameters
population_mean = 50
population_std = 10

# CLT parameters
sample_size = 30
num_samples = 1000

# Generate sample means
sample_means = []

for _ in range(num_samples):
    sample = np.random.normal(population_mean, population_std, sample_size)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True)
plt.xlabel("Sample Mean")
plt.ylabel("Probability Density")
plt.title("Verification of Central Limit Theorem")
plt.show()
```

12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_standard_normal():
    """
    Calculate and plot the standard normal distribution (mean = 0, std = 1)
    """
    # Generate x values
    x = np.linspace(-4, 4, 1000)

    # Standard normal PDF formula
    pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

    # Plot the distribution
    plt.plot(x, pdf)
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.title("Standard Normal Distribution (Mean = 0, Std = 1)"
```

13. Generate random variables and calculate their corresponding probabilities using the binomial distribution

```python
import numpy as np
import math

def binomial_pmf(k, n, p):
    """
    Calculate Binomial probability P(X = k)
    """
    return math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
samples = 10  # number of random variables to generate

# Generate binomial random variables
random_vars = np.random.binomial(n, p, samples)

print("Random Variable (X)  ->  Probability P(X)")
print("------------------------------------------")

# Calculate and display probabilities
for x in random_vars:
    prob = binomial_pmf(x, n, p)
    print(f"{x:>10}            ->  {prob:.4f}")
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution5

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_score(x, mean, std):
    """
    Calculate Z-score for a data point
    """
    return (x - mean) / std

# Given data
data_point = 70
mean = 60
std_dev = 5

# Calculate Z-score
z = calculate_z_score(data_point, mean, std_dev)
print("Z-score:", z)

# Generate standard normal distribution
z_values = np.linspace(-4, 4, 1000)
pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * z_values**2)

# Plot standard normal distribution
plt.plot(z_values, pdf)
plt.axvline(z, linestyle='--')
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.title("Z-score Compared to Standard Normal Distribution")
plt.show()
```
